```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
```

```python
df_x = df.drop('MEDV', axis=1)
df_y = df['MEDV']
```

```python
df_x.head()
```

```python
df_y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
mse = mean_squared_error(y_test, lr_pred)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
```

```python
# from keras.api._v2.keras import activations
model = Sequential()
# model.add(Dense(128, activation='relu', input_shape=(X_train.shape)))
model.add(Dense(128,activation='relu', input_shape=(13,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
history = model.fit(X_train, y_train, epochs=200, verbose=2, batch_size=4, validation_data=(X_test, y_test))
```

```python
dnn_pred = model.predict(X_test)
```

```python
mse_dnn = mean_squared_error(y_test, dnn_pred)
```
